```python
import os 
import pandas as pd
```

```python
path ="../Extract/Country_Weather.csv"

weather_original = pd.read_csv(path)
weather_original.head()
```

```python
# Drop unneeded columns
weather_df = weather_original[['Country', 'Precipitation', 'Temp Avg', 'Temp Max', 'Temp Min']]

# Renaming Countries
weather_df.replace('Bosnia', 'Bosnia and Herzegovina', inplace=True)
```

```python
# Changing the date from days to months
date_format = pd.to_datetime(weather_original['Date'], format='%m/%d/%Y').dt.strftime('%m/%Y')

# Adding new date column to df
weather_df['Date'] = date_format
weather_df
```

```python
# Individual dfs to hold values while calculating monthly values
avg_weather = weather_df[['Country','Date', 'Temp Avg']]
min_weather = weather_df[['Country', 'Temp Min']]
max_weather = weather_df[['Country','Temp Max']]
precip_weather = weather_df[['Country', 'Precipitation']]

# Find monthly precip values and min, max and avg temps
avg_weather_df1 = avg_weather.groupby(['Country', 'Date']).mean()
min_weather_df1 = min_weather.groupby(['Country']).min()
max_weather_df1 = max_weather.groupby(['Country']).max()
precip_weather_df1 = precip_weather.groupby(['Country']).sum()


# Reset index for dfs
avg_weather_df = avg_weather_df1.reset_index()
min_weather_df = min_weather_df1.reset_index()
max_weather_df = max_weather_df1.reset_index()
precip_weather_df = precip_weather_df1.reset_index()


print(avg_weather_df)
print(min_weather_df)
print(max_weather_df)
print(precip_weather_df)
```

```python
# Merging monthly dfs into one
partial_df = pd.merge(avg_weather_df, min_weather_df, on="Country")
other_df = pd.merge(max_weather_df, precip_weather_df, on="Country")
covid_weather_df1 = pd.merge(partial_df, other_df, on="Country")

# Round columns to 2 digits
covid_weather_df = covid_weather_df1.round(2)
covid_weather_df.head()
```

```python
country_list = covid_weather_df['Country'].unique().tolist()
len(country_list)
```

```python
country_df1 = pd.DataFrame(country_list)
country_df = country_df1.reset_index()
country_df = country_df.rename(columns = {0:'Country'})
country_df
```

```python
# Save df to a csv file
country_df.to_csv ('country_list.csv', index = False, header = True)
```

```python
primary_key_df = pd.merge(country_df, covid_weather_df, on="Country")
primary_key_df = primary_key_df.reset_index()
primary_key_df = primary_key_df.rename(columns={'level_0':'Key', 'index':'Country Key'})
primary_key_df
```
