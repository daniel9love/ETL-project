```python
import os 
import pandas as pd
```

```python
path ="Time_Series_19-covid-Confirmed Cases.csv"

covid_df = pd.read_csv(path)
covid_df.head()
```

```python
# List of confirmed cases columns
column_list = covid_df.columns[4:49]

# Sum of all confirmed cases
covid_df['Confirmed Cases'] = covid_df[column_list].sum(axis=1)

covid_cleaned_df = covid_df[['Province/State','Country/Region', 'Lat','Long','Confirmed Cases']]
covid_cleaned_df
```

```python
covid_cleaned_df.to_csv ('covid19_cleaned', index = False, header = True)
```

```python
min_lat = covid_cleaned_df['Lat'].min()
max_lat = covid_cleaned_df['Lat'].max()
avg_lat = covid_cleaned_df['Lat'].mean()

print(f' Min Lat: {round(min_lat,2)}')
print(f' Max Lat: {round(max_lat,2)}')
print(f' Avg Lat: {round(avg_lat,2)}')
```

```python
total_confirmed = covid_cleaned_df['Confirmed Cases'].sum()
print(f'Total Confirmed Cases: {total_confirmed}')
```

```python
covid_spread = covid_cleaned_df.loc[(covid_cleaned_df["Country/Region"] != "Mainland China")]
covid_spread
```

```python
total_spread = covid_spread['Confirmed Cases'].sum()
print(f'Total Confirmed Cases Outside China: {total_spread}')
```

```python
min_lat = covid_spread['Lat'].min()
max_lat = covid_spread['Lat'].max()
avg_lat = covid_spread['Lat'].mean()

print(f' Min Lat: {round(min_lat,2)}')
print(f' Max Lat: {round(max_lat,2)}')
print(f' Avg Lat: {round(avg_lat,2)}')
```

```python
covid_north_hemi = covid_spread.loc[(covid_spread["Lat"] > 0)]
covid_south_hemi = covid_spread.loc[(covid_spread["Lat"] < 0)]

north_covid =  covid_north_hemi['Confirmed Cases'].sum()
south_covid =  covid_south_hemi['Confirmed Cases'].sum()

print(f'Northern Hemisphere Confirmed Cases (Outside of China): {north_covid}')
print(f'Southern Hemisphere Confirmed Cases: {south_covid}')
```
