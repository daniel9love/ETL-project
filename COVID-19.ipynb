```python
import os 
import pandas as pd
```

```python
path ="Time_Series_19-covid-Confirmed Cases.csv"

covid_df = pd.read_csv(path)
covid_df.head()
```

```python
# List of confirmed cases columns
column_list = covid_df.columns[4:49]
jan = covid_df.columns[4:14]
feb = covid_df.columns[15:43]
march = covid_df.columns[44:49]

# Sum of all confirmed cases
covid_df['Jan Confirmed Cases'] = covid_df[jan].sum(axis=1)
covid_df['Feb Confirmed Cases'] = covid_df[feb].sum(axis=1)
covid_df['March Confirmed Cases'] = covid_df[march].sum(axis=1)
covid_df['Total Confirmed Cases'] = covid_df[column_list].sum(axis=1)

covid_cleaned_df = covid_df[['Province/State','Country/Region', 'Lat','Long',
                             'Jan Confirmed Cases', 'Feb Confirmed Cases', 
                             'March Confirmed Cases','Total Confirmed Cases']]
covid_cleaned_df
```

```python
# Renaming Countries
covid_cleaned_df.replace('Mainland China', 'China', inplace=True)
covid_cleaned_df.replace('Hong Kong', 'China', inplace=True)
covid_cleaned_df.replace('Taiwan', 'China', inplace=True)
covid_cleaned_df.replace('UK', 'United Kingdom', inplace=True)
covid_cleaned_df.replace('US', 'United States', inplace=True)
covid_cleaned_df.replace('Andorra', 'France', inplace=True)
covid_cleaned_df.replace('Bhutan', 'China', inplace=True)
covid_cleaned_df.replace('Liechtenstein', 'Switzerland', inplace=True)
covid_cleaned_df.replace('Monaco', 'France', inplace=True)
covid_cleaned_df.replace('North Macedonia', 'Greece', inplace=True)
covid_cleaned_df.replace('Palestine', 'Israel', inplace=True)
covid_cleaned_df.replace('Saint Barthelemy', 'Others', inplace=True)
covid_cleaned_df.replace('San Marino', 'Italy', inplace=True)
```

```python
# Dropping unneeded columns
covid19_cleaned = covid_cleaned_df[['Country/Region','Jan Confirmed Cases', 
                    'Feb Confirmed Cases', 'March Confirmed Cases','Total Confirmed Cases']]

# Grouping counts by country, resetting index, renaming columns
covid19_cleaned_df = covid19_cleaned.groupby(['Country/Region']).sum()
covid19_cleaned_df = covid19_cleaned_df.reset_index()
covid19_cleaned_df = covid19_cleaned_df.rename(columns = {'Country/Region':'Country'})
 
covid19_cleaned_df 
```

```python
# Save df to a csv file
covid19_cleaned_df.to_csv('covid19_cleaned.csv', index = False, header = True)
```

```python
# Finding min, max and avg latitude
min_lat = covid_cleaned_df['Lat'].min()
max_lat = covid_cleaned_df['Lat'].max()
avg_lat = covid_cleaned_df['Lat'].mean()

print(f' Min Lat: {round(min_lat,2)}')
print(f' Max Lat: {round(max_lat,2)}')
print(f' Avg Lat: {round(avg_lat,2)}')
```

```python
# Calculate total confirmed cases
total_confirmed = covid_cleaned_df['Total Confirmed Cases'].sum()
print(f'Total Confirmed Cases: {total_confirmed}')
```

```python
# Countries covid 19 has spread to
covid_spread = covid_cleaned_df.loc[(covid_cleaned_df["Country/Region"] != "Mainland China")]
covid_spread
```

```python
# Calculating confirmed cases outside of China
total_spread = covid_spread['Total Confirmed Cases'].sum()
print(f'Total Confirmed Cases Outside China: {total_spread}')
```

```python
# Calculating confirmed cases in Northern and Southern hemisphere
covid_north_hemi = covid_spread.loc[(covid_spread["Lat"] > 0)]
covid_south_hemi = covid_spread.loc[(covid_spread["Lat"] < 0)]

north_covid =  covid_north_hemi['Total Confirmed Cases'].sum()
south_covid =  covid_south_hemi['Total Confirmed Cases'].sum()

print(f'Northern Hemisphere Confirmed Cases (Outside of China): {north_covid}')
print(f'Southern Hemisphere Confirmed Cases: {south_covid}')
```
